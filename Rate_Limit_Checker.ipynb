```python
import requests
from datetime import datetime

def fetch_with_rate_limit_info(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("Request successful.")
            return response
        elif response.status_code == 429:  # Rate limit hit
            print("Rate limit hit.")
            retry_after = response.headers.get("Retry-After")
            reset_time = response.headers.get("X-RateLimit-Reset")

            if retry_after:
                print(f"Retry after: {retry_after} seconds.")
            if reset_time:
                reset_datetime = datetime.fromtimestamp(int(reset_time))
                print(f"Rate limit resets at: {reset_datetime}")

        else:
            print(f"Unexpected status code: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")

# Example usage
test_url = "https://fbref.com/en/comps/9/Premier-League-Stats"  # Replace with a real API endpoint if needed
fetch_with_rate_limit_info(test_url)
```
